```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy.random import seed, randint
seed(100)

# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=False, sharey=False, dpi=200)

# Define the colors and markers to use
colors = {0:'g', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

# Plot each axes
for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('Ax: ' + str(i))
    ax.yaxis.set_ticks_position('none')

plt.suptitle('Four Subplots in One Figure', verticalalignment='bottom', fontsize=16)    
plt.tight_layout()
# plt.show()
```

### Go through and play  with the code above to try answer the questions below:

- What do you think `sharex` and `sharey` do?
   - When sharex & sharey are false, they will not share their x / y axes (scale)
   - when sharex & sharey are true, the graph will share the same scales

- What does the `dpi` argument control?
   - it zooms the graph, smaller or larger depending on the int value

- What does `numpy.ravel()` do, and why do they call it here?
   - .ravel() merges lists into one array
      - exmaple array([0,1,2],[3,4,5]) => array([0,1,2,3,4,5])

- What does `yaxis.set_ticks_position()` do?
   - it sets the position or the y axis ticks to the desired input (left, right, both, none)

- How do they use the `colors` and `markers` dictionaries?
   - iterates through the 'colors' and 'markers' dictionaries with 'i' being the index value

```python
from numpy.random import seed, randint
seed(100)

x = sorted(randint(0,10,10))
x2 = sorted(randint(0,20,10))
y = sorted(randint(0,10,10))
y2 = sorted(randint(0,20,10))
```

```python
# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=False, sharey=False, dpi=120)

# Define the colors and markers to use
colors = {0:'g', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}
plt.suptitle('Four Subplots in One Figure', verticalalignment='bottom', fontsize=16)    

axes[0][0].set_title('Plot 1')
axes[0][0].plot(x,y, color=colors[1])

axes[0][1].set_title('Plot 2')
axes[0][1].scatter(x2,y2, facecolors='none', color=colors[2])

axes[1][0].set_title('Plot 3')
axes[1][0].plot(x,y, color=colors[1])
axes[1][0].scatter(x2,y2, facecolors='none', color=colors[2])

axes[1][1].set_title('Plot 4')
axes[1][1].plot(x,y, color=colors[1])
ax4 = axes[1][1].twinx()
ax4.scatter(x2,y2, facecolors='none', color=colors[2])

fig.savefig('answer_guide.png')

# Plot each axes
# plt.tight_layout()
# plt.show()
```
